Correlation Contrast

```python
# Filter data for "color" condition only (DT == 'Contrast')
contrast_data = df[df['DT'] == 'Contrast'].copy()

print(f"Total rows in original dataset: {len(df)}")
print(f"Rows with DT='Contrast': {len(contrast_data)}")
print("\nContrast condition data:")
print(contrast_data[['DT', 'OC', 'visualSearchTime', 'ErrorCount']].head())
```

```python
# Map OC categories to numerical values for correlation analysis
oc_mapping = {'low': 1, 'medium': 2, 'high': 3}
contrast_data['OC_numeric'] = contrast_data['OC'].map(oc_mapping)

# Check the mapping
print("OC value distribution:")
print(contrast_data['OC'].value_counts())
print("\nMapped numeric values:")
print(contrast_data['OC_numeric'].value_counts())
```

```python
# Calculate correlation
correlation_results = pg.corr(contrast_data['OC_numeric'], 
                              contrast_data['visualSearchTime'],
                              method='pearson')  

print("Pearson correlation results:")
print(correlation_results)
```

```python
def summarizeDF(df, factors, measure):

    summary = df.groupby(factors)[measure].agg(
        Mean='mean',
        SD='std',
        N='count'
    ).reset_index()

    # erreur standard
    summary['SE'] = summary['SD'] / np.sqrt(summary['N'])

    # intervalle d'erreur (SE)
    summary['errory_hi'] = summary['SE']
    summary['errory_lo'] = summary['SE']

    return summary
```

```python
import pandas as pd

# Définir l'ordre souhaité
df_lighting['OC'] = pd.Categorical(
    df_lighting['OC'],
    categories=['low', 'medium', 'high'],
    ordered=True
)

# Puis refaire ton résumé
stats_lighting = summarizeDF(df_lighting, ['OC'], 'visualSearchTime')
print(stats_lighting)
```

```python
import plotly.express as px

nice_color_palette = ['#66c2a5', '#fc8d62', '#8da0cb']

fig = px.bar(
    stats_lighting,
    x='OC',
    y='Mean',
    error_y='errory_hi',
    error_y_minus='errory_lo',
    color='OC',
    color_discrete_sequence=nice_color_palette
)

fig.update_layout({
    'plot_bgcolor': 'rgba(0,0,0,0)',
    'title': 'Visual Search Time by Object Count (Lighting Condition)',
    'xaxis_title': 'Object Count',
    'yaxis_title': 'Mean Visual Search Time'
})

fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')

fig.show()
```
